```python
import pandas as pd
import numpy as nps
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
df_gdp= pd.read_csv('GDP.csv')
```

```python
df_gdp.head()
```

```python
df_gdp.tail()
```

```python
df_gdp_all = df_gdp.dropna()
```

```python
df_gdp_all
```

```python
df = df_gdp_all[(df_gdp_all['Country Name']=='India')| (df_gdp_all['Country Name']=='Pakistan')]
df
```

```python
df= df.set_index('Country Name')
```

```python
df1=df.T

df1
```

```python
df1.plot()
plt.xlabel('Years')
plt.ylabel('GDP in US Dollars $')
plt.title('India & Pakistan GDP in Trillion US$ 1960-2018')
plt.show()
```
